# Snippet - Regressão Linear pelos Mínimos Quadrados Ordinários
A regressão linear ajusta uma reta com coeficientes que minimizam a soma residual dos quadrados entre os pontos de dados observados e os previstos pela reta.

A fórmula do modelo é descrita como: $\hat y(w,x) = w_0 + w_1x_1 + ... + w_px_p$

O vetor dos coeficientes $w_n$ é acessado pelo atributo `coef_`, enquanto que o intercepto $w_0$ é acessado por `intercept_`.

A função objetivo é descrita como: $min||Xw - y||^2$

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Data
data = {
    "x1": [1,2,3,4,5],
    "x2": [1,2,3,4,5],
    "x3": [1,2,3,4,5],
    "x4": [1,2,3,4,5],
    "x5": [1,2,3,4,5],
    "y": [1,2,3,4,5]
}

df = pd.DataFrame(data)
```

```python
# Fit
lg = linear_model.LinearRegression()
lg.fit(df[["x1","x2","x3","x4","x5"]],df["y"])
```

```python
# Attributes
print(f'Intercept: {lg.intercept_}\nCoeficients:{lg.coef_}')
```

```python
# Prep
X = df[["x1","x2","x3","x4","x5"]]
y = df["y"]
y_pred = lg.intercept_ + np.dot(X, lg.coef_)

# Viz
plt.scatter(range(len(y)), y, color='blue', label='Real y')
plt.plot(range(len(y_pred)), y_pred, color='red', label='Regressor line')
plt.xlabel('Índice')
plt.ylabel('y')
plt.legend()
plt.title('Real y vs. Pred y')
plt.show()
```
