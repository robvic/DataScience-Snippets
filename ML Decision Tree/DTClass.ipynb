# Snippet - Árvore de Decisão - Classificador
Árvores de decisão (DTs) são um método de aprendizagem supervisionado não paramétrico usado para classificação e regressão. O objetivo é criar um modelo que preveja o valor de uma variável alvo, aprendendo regras de decisão simples inferidas a partir dos recursos dos dados. Uma árvore pode ser vista como uma aproximação constante por partes.

No exemplo abaixo, as árvores de decisão aprendem com os dados a aproximar uma curva senoidal com um conjunto de regras de decisão se-então-senão. Quanto mais profunda a árvore, mais complexas serão as regras de decisão e mais ajustado será o modelo.

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import load_iris
```

```python
# Data
data = load_iris()
X, y = data.data, data.target
```

```python
# Fit
clf = tree.DecisionTreeClassifier()
clf.fit(X,y)
```
